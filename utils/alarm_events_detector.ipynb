```python
import pandas as pd



def procesar_eventos(eventos, umbral_activacion=7, tolerancia=1):
    alarma_activada_prueba = 0
    contador_ladridos_seguidos_prueba = 0
    contador_no_es_ladrido_prueba = 0
    contador_eventos_prueba = 0
    escucha_activa_prueba = False
    alarma_eventos_prueba = []
    
    ultimo_evento_no_es_ladrido = False

    for evento in eventos:
        if evento == ' ESCUCHA ACTIVA INICIADA':
            escucha_activa_prueba = True
            contador_ladridos_seguidos_prueba = 0
            contador_no_es_ladrido_prueba = 0
            contador_eventos_prueba = 0
            ultimo_evento_no_es_ladrido = False
        elif evento == ' ESCUCHA ACTIVA DESACTIVADA':
            escucha_activa_prueba = False
        elif escucha_activa_prueba:
            if evento == ' ladrido despues de escucha':
                contador_ladridos_seguidos_prueba += 1
                contador_eventos_prueba += 1
                ultimo_evento_no_es_ladrido = False
            elif evento == ' No es ladrido':
                if contador_no_es_ladrido_prueba <= tolerancia:  # Tolerancia a más de un evento "No es ladrido"
                    contador_no_es_ladrido_prueba += 1
                else:
                    contador_ladridos_seguidos_prueba = 0
                    contador_no_es_ladrido_prueba = 0
                    contador_eventos_prueba = 0

            # Si llegamos al umbral de activación, contamos una alarma
            if contador_ladridos_seguidos_prueba == umbral_activacion:
                alarma_activada_prueba += 1
                alarma_eventos_prueba.append(f'Alarma activada después de {contador_eventos_prueba} eventos')
                # Reiniciamos el contador de ladridos después de activar la alarma
                contador_ladridos_seguidos_prueba = 0

    return {
        'Total de alarmas activadas': alarma_activada_prueba,
        'Eventos de activación de alarma': alarma_eventos_prueba
    }
def buscar_secuencias_entre_escucha(eventos):
    secuencias = []
    secuencia_actual = []
    escucha_activa = False
    
    for evento in eventos:
        if evento == ' ESCUCHA ACTIVA INICIADA':
            escucha_activa = True
            secuencia_actual = [evento]
        elif evento == ' ESCUCHA ACTIVA DESACTIVADA' and escucha_activa:
            secuencia_actual.append(evento)
            secuencias.append(secuencia_actual)
            escucha_activa = False
        elif escucha_activa:
            secuencia_actual.append(evento)
    
    return secuencias

# Load the Excel file
file_path = 'resultados_experimentos_tlaxcala/resultados/resultados.xlsx'
sheet_data = pd.read_excel(file_path, sheet_name='experimento_3_r_1_archi')


eventos_excel = sheet_data['Unnamed: 3'].dropna().reset_index(drop=True)
resultado_excel = procesar_eventos(eventos_excel)
secuencias_encontradas = buscar_secuencias_entre_escucha(eventos_excel)


for secuencia in secuencias_encontradas:
    for i in secuencia:
        # print(i)
        pass
    print(procesar_eventos(secuencia))
```
